```python
import pandas as pd
```

```python

# Load the Wikipedia page
url = "https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table"
tables = pd.read_html(url)

# Assuming the third table is the one with the all-time medal data
medal_table = tables[2]  # Adjust this index if necessary

pd.set_option('display.max_rows', 100)
medal_table
```

```python
name_with_country_codes = medal_table["Team (IOC code)"]

def get_country_name(country):
    return country[0:country.find("(")].strip()


country_names = name_with_country_codes.map(get_country_name)
```
